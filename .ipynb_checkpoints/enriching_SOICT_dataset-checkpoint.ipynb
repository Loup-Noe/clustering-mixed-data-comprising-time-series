```python
def plot_helper(l_series, barycenter):
    # plot all points of the data set
    for series in l_series:
        plt.plot(series.ravel(), "k-", alpha=.2)
    # plot the given barycenter of them
    plt.plot(barycenter.ravel(), "r-", linewidth=2)
```

```python
df = df_energy

# Perform Fast Fourier Transform (FFT) on each column
fft_df = df.apply(lambda x: np.fft.fft(x))

# Calculate amplitudes and phases
amplitudes = fft_df.applymap(np.abs)
phases = fft_df.applymap(np.angle)

# Get the frequency information
num_samples = len(df)
freqs = np.fft.fftfreq(num_samples)

# Find the index of the dominant frequency component (excluding the zero-frequency component)
dominant_component_idx = amplitudes.iloc[1:].idxmax()

# Extract features for each column
features = []
for col in df.columns:
    dominant_amplitude = amplitudes.at[dominant_component_idx[col], col]
    dominant_phase = phases.at[dominant_component_idx[col], col]
    dominant_frequency = freqs[int(dominant_component_idx[col])]

    features.append((col, dominant_amplitude, dominant_phase, dominant_frequency))

# Convert features to a DataFrame
features_df = pd.DataFrame(features, columns=['Column', 'Amplitude', 'Phase', 'Frequency'])
```
